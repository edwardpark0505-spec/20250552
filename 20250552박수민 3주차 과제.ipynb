```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([1, 2, 3, 4, 5, 6, 7])
y = np.array([78.4, 78.6, 78.5, 78.9, 78.2, 79.4, 78.6])
plt.plot(x, y, color='r',marker='o')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace (-5, 5, 300)
yl = x ** 2
y2 = (x - 2) ** 2
plt.plot(x, yl, color = "r")
plt.plot(x, y2, color = "k", linestyle = "--")
plt.grid()
plt.show()
```

```python

# 6. 데이터 생성
x = np.linspace(-5, 5)
y1 = 2 * x
y2 = 5 * x

# 7. 그래프 그리기
plt.plot(x, y1, label="y1 값", color='r')
plt.plot(x, y2, label="y2 값", linestyle="dashed", color='k')

# 8. 축, 제목, 범례, 그리드 설정
plt.xlabel("X-축", size=14)
plt.ylabel("Y-축", size=14)
plt.title("그래프 제목")
plt.grid()
plt.legend()

# 9. 그래프 출력
plt.show()
```
